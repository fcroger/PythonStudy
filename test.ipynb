```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
from pandas import Series, DataFrame

def random_color():
    colors = '0123456789ABCDEF'
    num = '#'
    for i in range(6):
        num += random.choice(colors)
    return num

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus']=False
plt.figure(figsize=(16,10),dpi=80)
plt.title("NNO4A")

df=pd.read_csv('data/inclinometer_data.csv')
df_group=df.groupby('')

plt.grid()
values =[-20,-15,-10,-5,0,5,10,15,20]

for i in range(1,46):
    
    rc=random_color()
    
    if(i<10):
        sensor_no = 'RT2101001582.L1_SW_0' + str(i)
    else:
        sensor_no = 'RT2101001582.L1_SW_' + str(i)

    df=df_group.get_group(sensor_no)
    x=df['dispsx']
    y=df['dispsy']
    #x = np.random.randint(-5,10,10)
    #y = np.random.randint(0,80,10)
    #y.sort()
    plt.plot(x,y, color=rc, linestyle='-',linewidth=3,label=sensor_no) 
plt.xticks(range(-20,25,5))
plt.yticks(range(0,90,10))
ax = plt.gca()                                 #
ax.xaxis.set_ticks_position('top')   #X
plt.legend(loc='center right',bbox_to_anchor=(-0.02,0.5))
ax.invert_yaxis()                            #Y
plt.show()  
```
